```python
import geopandas
import pandas as pd

original = geopandas.read_file("/content/dammam_districts_with_indicators.geojson")
```

```python
df =  original[['district_id', 'name_en', 'name_ar', 'population_density_index', 'geometry']]
```

```python
df['particulate_matter'] = 15
```

```python
df['temperature'] = 24
```

```python
df['ROI_temperature'] = (df['temperature'] * df['population_density_index']) / 25
```

```python
df['ROI_particulate_matter'] = (df['particulate_matter'] * df['population_density_index']) / 15
```

```python
df = df[['district_id', 'name_en', 'name_ar', 'population_density_index', 'particulate_matter', 'temperature', 'ROI_temperature', 'ROI_particulate_matter','geometry']]
```
